# Listas vs Numpy array usando. un dataset de vinos 
Ahora que sabemos como crear numpy arrays, podemos cargar un dataset en una lista de listas y empalmar con conocimientos exitentes de listas.

Vamos a utilizar un dataset de datos numericos de vinos, para determinar la calidad del vino teniendo en cuenta las siguientes variables:
* fixed acidity
* volatile acidity
* citric acid
* residual sugar
* chlorides
* free sulfur dioxide
* total sulfur dioxide
* density
* pH
* sulphates
* alcohol

````python
import csv
import numpy as np

# Leemos nuestro dataset como una lista de listas
f = open("files/winequality-white.csv", "r")
wine_quality = list(csv.reader(f, delimiter=';'))

# Eliminamos el encabezado
wine_quality = wine_quality[1:]

# convertimos todos los valores a float, ya que estan en str
lol_wines = []
for row in wine_quality:
    converted_row = []
    for item in row:
        converted_row.append(float(item))
    lol_wines.append(converted_row)

# Put in a numpy array
wines = np.array(lol_wines)
````

# Vectorizaci贸n
Ya que notamos que seleccionar los datos con NumPy es m谩s sencillo, vamos a aprender que hacer otras operaciones sobre los datos tambi茅n es m谩s eficiente, ya que evitamos ejecutar una acci贸n cada vez que iteramos en el for loop,  y solamente efectuamos la operaci贸n una sola vez.

### Operaciones con vectorizaci贸n:
Son operaciones que se realizan entre vectores, o ndarrays de la misma dimension y forma (shape)
* **Suma:** vector_a + vector_b
* **Resta:** vector_a - vector_b
* **Multiplicaci贸n:** vector_a * vector_b (no es el producto punto)
* **Divisi贸n:** vector_a / vector_b

*Para realizar operaciones, las columnas (o vectores) deben tener la misma dimensi贸n, es decir, no puedo sumar un vector de 10 columnas con otro vector de 20 columnas, ya que se emparejan linea por l铆nea.

La operaci贸n s贸lo se hace una vez!


Para ilustrar el ejemplo, vamos a sumar dos columnas del dataset
###  Con lista de listas


````python
total_quality_alcohol = []

alcohol = [-2]
ph = lol_wines[-4]

for row in lol_wines:
    quality = row[-1]
    alcohol = row[-2]
    quality_alcohol = quality + alcohol
    total_quality_alcohol.append(quality_alcohol)
    
total_quality_alcohol
````
